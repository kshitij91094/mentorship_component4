### Calculating the average Sales Numbers and Attrition Numbers grouped by 'recruiting_source'


```python
# Grouping by 'recruiting_source' to get average sales_quota_pct and attrition
average_sales_attrition = rdata.groupby('recruiting_source').agg({
    'sales_quota_pct': 'mean',
    'attrition': 'mean'
}).reset_index()

# Displaying the results to check the average Sales and Attrition numbers
average_sales_attrition
```

```python
# Import necessary libraries
from plotnine import ggplot, aes, geom_bar, theme_minimal, labs

# Grouping the data by 'recruiting_source' and calculating the average attrition
average_attrition = rdata.groupby('recruiting_source')['attrition'].mean().reset_index()

# Create a bar plot to visualize attrition by recruiting source
plot = (ggplot(average_attrition, aes(x='recruiting_source', y='attrition'))
        + geom_bar(stat='identity', fill='skyblue')
        + theme_minimal()
        + labs(title='Attrition by Recruiting Source', x='Recruiting Source', y='Average Attrition'))

# Display the plot
plot.show()
```

```python
# Grouping the data by 'recruiting_source' and calculating the average sales
average_sales = rdata.groupby('recruiting_source')['sales_quota_pct'].mean().reset_index()

# Create a bar plot to visualize sales performance by recruiting source
plot = (ggplot(average_sales, aes(x='recruiting_source', y='sales_quota_pct'))
        + geom_bar(stat='identity', fill='lightgreen')
        + theme_minimal()
        + labs(title='Sales Performance by Recruiting Source', x='Recruiting Source', y='Average Sales Quota (%)'))

# Display the plot
plot.show()
```
